```python
# test_code_agent.py
from code_agent import Code  # 假设你的agent类在code_agent.py中

def test_code_agent():
    # 创建Code Agent实例
    code_agent = Code()
    
    # 测试不同类型的代码生成请求
    test_cases = [
        "Create a function to clean a pandas DataFrame"
    ]
    
    for case in test_cases:
        try:
            # 调用run方法
            generated_code = code_agent.run(case)
            
            # 打印生成的代码
            print(f"\n--- Test Case: {case} ---")
            print(generated_code)
            
            # 可选：简单的代码有效性检查
            assert generated_code is not None, "Generated code should not be None"
            assert len(generated_code.strip()) > 0, "Generated code should not be empty"
        
        except Exception as e:
            print(f"Error processing case '{case}': {e}")
            raise

# 如果直接运行这个脚本
if __name__ == "__main__":
    test_code_agent()
```
